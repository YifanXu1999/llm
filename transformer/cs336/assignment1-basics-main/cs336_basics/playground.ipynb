```python
special_tokens = ["<|endoftext|>", "<|endoftext|2>"]


import regex as re

x = 'abc <|endoftext|> efg'

print(re.split('|'.join(special_tokens), x))
```

```python

escaped_tokens = [re.escape(token) for token in special_tokens]
# Create regex pattern with capture groups to keep the tokens
pattern = f'({"|".join(escaped_tokens)})'
# Split and keep the delimiters
print(pattern)

pattern_2 = f'{"|".join(special_tokens)}'
print(pattern_2)
```
